# 1. 对数据的content进行字标志，分别是theme和sentiment_word。
- /B: 词的开头
- /M: 词的中间
- /E: 词的结尾
- /S: 单独成词
- /N: 不属于以上四类

```python
import pandas as pd
import numpy as np
data = pd.read_csv("./data/test.csv", sep='\t')
data
```

```python

import re
def tagging(content, words):
    index_dict = {}
    word_list = words.split(';')
    for word in word_list:
        if word != "NULL" and word != "":
            location = [m.start() for m in re.finditer(word, content)]
            for pos in location:
                index_dict[pos] = len(word)
        else:
            pass
    keys = list(index_dict.keys())
    keys = sorted(keys)
    pos = 0
    end = len(content) - 1
    i = 0
    res = []
    while(i <= end):
        if pos >= len(keys) or i != keys[pos]:
            res.append(content[i] + "/N")
            i = i + 1
        else:
            length = index_dict[keys[pos]]
            if length == 1:
                res.append(content[i] + "/S")
            else:
                for j in range(length):
                    if j == 0:
                        res.append(content[i + j] + "/B")
                    elif j == length - 1:
                        res.append(content[i + j] + "/E")
                    else:
                        res.append(content[i + j] + "/M")
            i = i + length
            pos = pos + 1
    s = res[0]
    for i in range(1, len(res)):
        s = s + ' ' + res[i]
    s = s + '\n'
    return s

for index, row in data.iterrows():
    with open('./data/theme_word_tagging_test.csv', 'a') as f:
        res = tagging(row["content"], row["theme"])
        f.write(res)
```
